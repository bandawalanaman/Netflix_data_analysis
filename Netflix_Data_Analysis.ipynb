```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/netflix_titles.csv')

df_clean = df.dropna().copy()
sns.countplot(x='type', data=df_clean)  # Count plot: Movies vs TV Shows

plt.title('Count of Movies vs TV Shows on Netflix')
plt.xlabel('Type of Content')
plt.ylabel('Number of Titles')
plt.show()

# Top 10 countries with most titles
top_countries = df_clean['country'].value_counts().head(10)
top_countries.plot(kind='bar', color='orange')
plt.title('Top 10 Countries with Most Netflix Titles')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.show()

# Top 10 genres
top_genres = df_clean['listed_in'].value_counts().head(10)
top_genres.plot(kind='barh', color='green')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Genre')
plt.show()

df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df_clean = df.dropna(subset=['date_added']).copy() #
df_clean.loc[:, 'year_added'] = df_clean['date_added'].dt.year

# Titles added per year
df_clean['year_added'].value_counts().sort_index().plot(kind='bar')
plt.title('Content Added on Netflix Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.show()
#top 10 directors
top_directors = df_clean['director'].value_counts().head(10)
top_directors.plot(kind='barh', color='purple')
plt.title('Top 10 Directors with Most Titles')
plt.xlabel('Number of Titles')
plt.ylabel('Director')
plt.show()

```
